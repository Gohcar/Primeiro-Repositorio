```python
#importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#criando dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualiando as primeiras 5 linhas
df.head()
```

```python
#Quantidade de linhas e coluas
df.shape
```

```python
#verificando os tipos de dados
df.dtypes
```

```python
#Qual foi a Receita total?
df["Valor Venda"].sum()
```

```python
#qual custo total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
#qual custo total?
round(df["custo"].sum(), 2)
```

```python
#qual o lucro?
df["Lucro"] = df["Valor Venda"] - df["custo"]
df["Lucro"].sum()
```

```python
#Qual tempo de envio do produto?
df["tempo entrega"] = df["Data Venda"] - df["Data Envio"]
```

```python
df.head()
```

```python
#Extraindo apenas os dias
df["tempo entrega"] = (df["Data Envio"]- df["Data Venda"] ).dt.days
df.head()
```

```python
df["tempo entrega"].dtype
```

```python
#Media de tempo de envio por marca
df.groupby("Marca")["tempo entrega"].mean()
```

```python
#vericando valores em branco
df.isnull().sum()
```

```python
#Encontrar Lucro por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = "{:20,.2f}".format
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual total de produtos vendidos por produto
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True)
```

```python
#Graficoc
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```
